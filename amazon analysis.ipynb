```python
import pandas as pd 
import numpy as np 
```

```python
data=data.drop_duplicates()
```

```python
data.head()
```

```python
data.drop(["description"],axis=1,inplace=True)
```

```python
data.head()
```

```python
data['director'].fillna("unavailable",inplace=True)
data['cast'].fillna("unavailable",inplace=True)
data['country']=data['country'].fillna(data['country'].mode()[0])
data['date_added']=data['date_added'].ffill()
data['rating']=data['rating'].fillna(data['rating'].mode()[0]) 
```

```python
data.isna().sum()
```

```python
data.head()
```

```python
data['date_added']=pd.to_datetime(data['date_added'],format='%B %d, %Y')
```

```python
data.head()
```

```python
data['date_added']
```

```python
data['rating'].unique()
```

```python
data['rating']=data['rating'].replace(
    {
        "NOT_RATE":"UNRATED", 
        "AGES_16_":"16+", 
        "16":"16+", 
        "AGES_18_":"18+",
        "ALL_AGES":"ALL"
    }
)
```

```python
data['rating'].unique()
```

```python
data['type'].value_counts()
```

```python
data.rename(columns={"listed_in":"categories"},inplace=True)
```

```python
data.head()
```

```python
filterd_actor = data[data['cast'] !="unavailable"] 
top_actor = filterd_actor['cast'].value_counts().head(10).to_frame() 
top_actor
```

```python
data['country'].value_counts().to_frame()
```

```python
data['month']=data['date_added'].dt.month
data['year']=data['date_added'].dt.year
```
